```python
from numpy import*
import numpy as np
from numpy.linalg import*
import matplotlib.pyplot as plt
#import pyprocar
import subprocess

def Pattern(N,Q,R): 
    '''
    This function takes two files: data1 and data2.... N is the number of size, 
    Q is the Piezo label, R is the Piezo Site to view. Field is the Field you are viewing
    '''
    
    for num in range(0,2200,200):
        data1 = np.loadtxt('/Users/sholafolarin/Documents/Quench_300k/Q_0_300k/PIEZO'.format(num))
        data2 = np.loadtxt('/Users/sholafolarin/Documents/Quench_300k/Q_0_300k/VectFieldY'.format(num))
    



        M=10



        plt.rc('axes', linewidth=2)

        b = array(data1[:])
        a = array(data2[:])


        x = zeros((N*N))
        y = zeros((N*N))
        z = zeros((N*N))

        pc=10

        x[:] = a[N*N*M:N*N*(M+1),3]
        y[:] = a[N*N*M:N*N*(M+1),4]
        
        

        z[:] = b[N*N*M:N*N*(M+1),R]


    

        x1 = transpose(x.reshape(N,N))
        y1 = transpose(y.reshape(N,N))
        z1 = transpose(z.reshape(N,N))
#
#
        for i in arange(N):
            for j in arange(N):
                plt.scatter(i, j, s=1)
                plt.arrow(i, j, dx=(x1[i,j])*0.55, dy=(y1[i,j])*0.55, width=.06) 
        
# 
        l = arange(0, N, 1)
        m = arange(0,N,1)

        X, Y = meshgrid(l, m)  

        plot = plt.pcolormesh(l, m, z1, cmap='jet', shading='gouraud')


        clb = plt.colorbar()
        clb.ax.set_title(Q)

        plt.ylim(0,N-1)
        plt.xlim(0,N-1)
        plt.gca().set_xticklabels(['']*10)
        plt.gca().set_yticklabels(['']*10)
        plt.xlabel('X', fontsize = 20)
        plt.ylabel('Y',fontsize = 20)
        plt.tick_params(direction='in', pad=10)
        plt.tick_params('both', length=5, width=2)
        #plt.savefig(Q +'.jpeg', format='jpeg')
        plt.suptitle('Field --> {}Kv/cm, Quenched from 300K to 10K'.format(num,N,N,N))
        plt.savefig('e11_{}.jpeg'.format(num), format='jpeg', bbox_inches='tight')
        plt.show(block=True)
    
```
